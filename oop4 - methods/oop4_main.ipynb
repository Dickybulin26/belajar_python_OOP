### **OOP Kelas Terbuka #4 - Methods**

```python
class Hero: #* template
    #* class variabel
    jumlah_hero = 0

    def __init__(self, inputName, inputHealth, inputPower, inputArmor):
        #* instance varibel
        self.name = inputName
        self.health = inputHealth
        self.power = inputPower
        self.armor = inputArmor
        Hero.jumlah_hero += 1

    #* viod funtion, method tanpa function
    def siapa(self):
        print(f'namaku adalah {self.name}')

    #* method dengan argumen
    def healthUP(self,up):
        self.health += up

    #* method dengan return
    def getHealth(self):
        return self.health
```

```python
hero1 = Hero('Sniper', 100, 10 ,5)
hero2 = Hero(' mario bros', 99, 5 ,10)
```

```python
print(hero1.__dict__)
print(hero2.__dict__)
```

```python
hero1.siapa()
hero1.healthUP(10)
```

```python
print(hero1.getHealth())
```
