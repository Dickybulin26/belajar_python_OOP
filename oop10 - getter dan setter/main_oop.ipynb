### **OOP Kelas Terbuka #10 - Getter & Setter**

```python
class Hero:

    #* private class variabel
    __jumlah = 0

    def __init__(self,name,health,armor):
        self.name = name
        self.__health = health
        self.__armor = armor
        # self.info = f'name {self.name}: \n\thealth: {self.__health}'

    @property
    def info(self):
      return f'name {self.name}: \n\thealth: {self.__health}'
    
    #* getter
    @property
    def armor(self):
      pass

    @armor.getter
    def armor(self):
      return self.__armor

    #* setter
    @armor.setter
    def armor(self,input):
      self.__armor = input

    #* deleter
    @armor.deleter
    def armor(self):
      print('armor di delet')
      self.__armor = None

sniper = Hero('sniper', 100,10)
```

```python
print('merubah info')
print(sniper.info)
sniper.name = 'dadang'
print(sniper.info)
```

```python
print('getter dan setter untuk __armor:')
print(sniper.armor)
print(sniper.__dict__)
sniper.armor = 50
print(sniper.armor)

print(sniper.__dict__)
```

```python
print('delet armor')
del sniper.armor
print(sniper.__dict__)
```
