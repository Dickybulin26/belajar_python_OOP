### **OOP Kelas Terbuka #21 - Abstract Class & Decorator**


```python
from abc import ABC,abstractmethod

class Button(ABC):

    def __init__(self,set_link):
        self.link = set_link

    @abstractmethod
    def click(self):
        pass

    @property
    @abstractmethod
    def link(self):
        pass

class pushButton(Button):
    
    def click(self):
        print(f'Go To: {self.link}')

    @Button.link.setter
    def link(self,input):
        self.__link = input


    @link.getter
    def link(self):
        return self.__link

tombol1 = pushButton('https://youtu.be/4U38PtnbMsk')
tombol1.click()
```
